```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_normal_dist(n, mean=0, std=1):
 data = np.random.normal(mean, std, n)
 plt.hist(data, bins=100, density=True)
 plt.title(f'Normal Distribution(n={n}, mean={mean}, std={std})')
 plt.xlabel('value')
 plt.ylabel('Frequency')
 plt.grid(True)
 plt.show()


```

#اجرای تابع با n=10

```python
plot_normal_dist(10)
```

#اجرای تابع با n=1000

```python
plot_normal_dist(1000)
```

#اجرای تابع با n=100000

```python
plot_normal_dist(100000)
```
